```python
from collections import defaultdict
dic = defaultdict(list)
dic['a'].append([1, 2, 3])
dic['b'].append([4, 5, 6])
print(dic)
```

```python
setb = defaultdict(set)
setb['a'].add((1, 1, 1))
print(setb)
```

```python
nums = [1,1,1,2,2,3]
```

```python
slow = 2
fast = 2

while fast < len(nums) - 1 :
    if nums[fast] != nums[slow]:
        nums[slow] = nums[fast]
        slow += 1
    fast +=1


```
