```python
# Find the key with the maximum value without using builtin max()
max_key_manual = None
max_value = None

for key in char:
    
    if max_value is None or char[key] > max_value:
        max_value = char[key]
        max_key_manual = key

print(f"Max key: {max_key_manual}, Max value: {max_value}")
```

```python
max_key = max(char, key=char.get)
print(max_key)
```

```python
# iterate through keys
for k in char:
    print("key:", k, "value:", char[k])

# iterate through keys explicitly
for k in char.keys():
    print("key:", k)

# iterate through values
for v in char.values():
    print("value:", v)

# iterate through key/value pairs
for k, v in char.items():
    print("key:", k, "value:", v)

# iterate with index
for i, (k, v) in enumerate(char.items()):
    print("index:", i, "key:", k, "value:", v)
```
