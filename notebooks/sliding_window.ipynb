```python
# Example: find all start indices of anagrams of p in s (sliding window)
# Uses a fixed-size char array (256) instead of a hashmap.
# Works in O(n + K) time (K=256 here), and is typically faster than a hashmap
# when alphabet is small/bounded (ASCII). Use hashmap if alphabet is large/unbounded.

def find_anagrams(s: str, p: str):
    if len(p) > len(s):
        return []
    count = [0] * 256  # you can reuse the notebook's char_array if you prefer
    for ch in p:
        count[ord(ch)] += 1

    need = len(p)
    res = []
    for i, ch in enumerate(s):
        o = ord(ch)
        if count[o] > 0:
            need -= 1
        count[o] -= 1

        if i >= len(p):
            left_o = ord(s[i - len(p)])
            count[left_o] += 1
            if count[left_o] > 0:
                need += 1

        if need == 0:
            res.append(i - len(p) + 1)
    return res

# Example usage
s = "cbaebabacd"
p = "abc"
print(find_anagrams(s, p))  # [0, 6]
```
