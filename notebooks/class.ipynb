```python
class car :

    car_count = 0
    car_wheels = 4
    car_gear = 6

    def __init__(self, model , company , engine, price):
        self.model = model
        self.company = company
        self.engine = engine
        self.price = price
        car.car_count += 1 


    @classmethod
    def car_gears(cls):
        return cls.car_gear
    
    
    @classmethod
    def total_cars(cls):
        return cls.car_count

    @classmethod
    def total_wheels(cls):
        return cls.car_wheels
    
    
    def is_expensive(self):
        if self.price > 10000000:
            return "Expensive Car"
        else : 
            return "Affordable Car"
        


    def show_details(self):
        return (f'Model: {self.model}, Company : {self.company}, Engine : {self.engine}')

    def change_engine(self,new_engine):
        self.engine = new_engine

    def update_company(self,updated_company):
        self.company = updated_company

    def __str__(self):
        return "This class create car objects"
        
```

```python
print(car.total_cars())
print(car.total_wheels())
print(car.car_gears())
```

```python
class Vehicle :
    def __init__(self, model , company):
        self.model = model 
        self.company = company

    def all_details(self):
        return f"Model : {self.model}, Company : {self.company}"
```

```python
class car(Vehicle):
    def __init__(self, model , company ,engine):
        super().__init__(model , company)
        self.engine = engine

    
```

```python
Maruti = car("zx", "suzuki","inline")
print(Maruti.company,Maruti.engine,Maruti.model)
```

```python
print(Maruti.all_details())
```

```python
class Bike(Vehicle):

    def __init__(self, model, company, mileage):
        super().__init__(model, company)
        self.mileage = mileage

    def all_details(self):
        return super().all_details() + " ," + f"Mileage: {self.mileage}"


```

```python
yamaha = Bike("R15", "Yamaha",23)
```

```python
print(yamaha.all_details())
```
